```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
df = pd.read_csv('/Users/gonuni/Desktop/College/CBB/HIV_LLMs/deploy_medical_LLM_evaluation/deploy_medical_llm_evaluation/evaluation_results/evaluation_results_merged.csv')

# Group by 'category_id' and 'model', and calculate the mean and std
grouped = df.groupby(['category_id', 'model']).agg({
    'overall_score_mean': 'mean',
    'overall_score_std': 'mean',
    'overall_score_wo_Q4&5_mean': 'mean',
    'overall_score_wo_Q4&5_std': 'mean'
}).reset_index()

# Get unique categories and models
categories = grouped['category_id'].unique()
models = grouped['model'].unique()
n_categories = len(categories)
n_models = len(models)

# Calculate bar heights and positions
bar_height = 0.8 / n_models  # Total bar height divided among models
y_positions = np.arange(n_categories)

# Create a figure with two subplots (1 row, 2 columns)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Plot 1: Overall scores
for i, model in enumerate(models):
    # Filter data for the current model
    model_data = grouped[grouped['model'] == model]
    # Map category IDs to y positions
    y_offsets = y_positions - 0.4 + i * bar_height + bar_height / 2
    # Get means and standard deviations
    means = model_data['overall_score_mean']
    stds = model_data['overall_score_std']
    # Plot horizontal bars
    ax1.barh(y_offsets, means, height=bar_height, xerr=stds, align='center', label=model, capsize=5)

ax1.set_yticks(y_positions)
ax1.set_yticklabels(categories)
ax1.set_xlabel('Overall Score')
ax1.set_ylabel('Category ID')
ax1.set_title('Overall Score Per Category')
ax1.legend(title='Model')
ax1.grid(axis='x', linestyle='--', alpha=0.7)

# Plot 2: Overall scores without Q4&5
for i, model in enumerate(models):
    # Filter data for the current model
    model_data = grouped[grouped['model'] == model]
    # Map category IDs to y positions
    y_offsets = y_positions - 0.4 + i * bar_height + bar_height / 2
    # Get means and standard deviations
    means = model_data['overall_score_wo_Q4&5_mean']
    stds = model_data['overall_score_wo_Q4&5_std']
    # Plot horizontal bars
    ax2.barh(y_offsets, means, height=bar_height, xerr=stds, align='center', label=model, capsize=5)

ax2.set_yticks(y_positions)
ax2.set_yticklabels([])  # Remove y-axis labels on the second plot
ax2.set_xlabel('Overall Score w/o Q4&5')
ax2.set_title('Overall Score Per Category without Q4&5')
ax2.legend(title='Model')
ax2.grid(axis='x', linestyle='--', alpha=0.7)

# Adjust the layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('../deploy_medical_llm_evaluation/evaluation_results/evaluation_results_merged.csv')

# Sort the dataframe by category_id, question_index, iteration_number, and model to maintain order
df_sorted = df.sort_values(['category_id', 'question_index', 'iteration_number', 'model'])

# Define the columns for scores
scores_columns = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5']

# For each category, plot the individual scores for Q1 to Q5 for each question_index, iteration, and model
for category_id in df_sorted['category_id'].unique():
    # Filter data for the current category
    category_data = df_sorted[df_sorted['category_id'] == category_id]
    
    # Loop over each score (Q1 to Q5) to keep separate plots for each question
    for score in scores_columns:
        plt.figure(figsize=(10, 6))
        plt.title(f'Scores for Category {category_id} - {score}')
        plt.ylabel('Score')
        
        # Plot each model as a separate bar color
        for model in category_data['model'].unique():
            model_data = category_data[category_data['model'] == model]
            # Pivot to get score values for the specific question
            model_data_plot = model_data.pivot_table(values=score, index=['question_index', 'num_iterations'])
            model_data_plot.plot(kind='bar', ax=plt.gca(), label=model)
        
        # Customize the plot
        plt.xticks(rotation=45, ha='right')
        plt.legend(title="Model", loc="upper right")
        plt.tight_layout()
        plt.show()
```

```python
### Plot 2: Horizontal Bar Plot of Q1 to Q5 Scores per Category
df = pd.read_csv('./evaluation_results_GPT4-score.csv')

# Use 'category_id' in groupings to get the mean
grouped_mean = df.groupby(['model', 'category_id']).mean()

# Reset the index so that 'category_id' is a regular column
grouped_mean = grouped_mean.reset_index()

# Sort categories for consistent plotting
grouped_mean = grouped_mean.sort_values('category_id')

# Select the required columns
score_columns = ['Q1_mean', 'Q2_mean', 'Q3_mean', 'Q4_mean', 'Q5_mean']
std_columns = ['Q1_std', 'Q2_std', 'Q3_std', 'Q4_std', 'Q5_std']

# Prepare data for plotting
data = grouped_mean[['category_id'] + score_columns + std_columns]

# Melt the DataFrame to long format for seaborn
melted_data = pd.melt(
    data,
    id_vars='category_id',
    value_vars=score_columns,
    var_name='Question',
    value_name='Score Mean'
)

# Extract the standard deviations and merge
std_data = pd.melt(
    data,
    id_vars='category_id',
    value_vars=std_columns,
    var_name='Question',
    value_name='Score Std'
)

# Clean up the 'Question' column to match
melted_data['Question'] = melted_data['Question'].str.replace('_mean', '')
std_data['Question'] = std_data['Question'].str.replace('_std', '')

# Merge the mean and std data
plot_data = pd.merge(melted_data, std_data, on=['category_id', 'Question'])

# Reset index to ensure alignment
plot_data = plot_data.reset_index(drop=True)

# Set up the plot
plt.figure(figsize=(12, 8))

# Create a color palette with 5 colors (matching the number of questions)
palette = sns.color_palette('pastel', n_colors=5)

# Plotting
ax = sns.barplot(
    data=plot_data,
    y='category_id',
    x='Score Mean',
    hue='Question',
    palette=palette,
    orient='h',
    errorbar=None  # Updated per FutureWarning
)

# Add error bars
# Get the positions of the bars
for (index, row), bar in zip(plot_data.iterrows(), ax.patches):
    # Get the x and y position of the bar
    width = bar.get_width()
    y = bar.get_y() + bar.get_height() / 2
    xerr = row['Score Std']
    # Draw the error bar
    ax.errorbar(
        x=width,
        y=y,
        xerr=xerr,
        fmt='none',
        ecolor='black',
        capsize=3
    )

# Labels and title
plt.xlabel('Score Mean')
plt.ylabel('Category ID')
plt.title('Llama 3.1 70B Clinical HIV Usage Evaluation: Scores of Q1 to Q5 per Category')

# Adjust legend
plt.legend(title='Question', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show grid lines for clarity
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.savefig('Llama3.1_Clinical_HIV_usage_Evaluation.jpg')
plt.show()
```

```python
# Set the seaborn style
sns.set(style='whitegrid')

# Sort categories for consistent plotting
df = df.sort_values(['category_id', 'question_index'])

# Prepare data for plotting
plot_data = df.copy()

# Convert 'question_index' to integer for sorting
plot_data['question_index'] = plot_data['question_index'].astype(int)

# Create a new column combining category and question index for labeling
plot_data['Category_Question'] = 'Cat ' + plot_data['category_id'].astype(str) + ' Q' + plot_data['question_index'].astype(str)

# Reset index to ensure correct order
plot_data = plot_data.reset_index(drop=True)

# Map category IDs to integers for coloring
category_list = plot_data['category_id'].unique()
category_map = {category: idx for idx, category in enumerate(category_list)}
plot_data['category_idx'] = plot_data['category_id'].map(category_map)

# Add a new y position to introduce gaps between categories
plot_data['y_pos'] = 0

current_y = 0
gap_size = 1  # Adjust gap size as needed

for category in category_list:
    category_indices = plot_data[plot_data['category_id'] == category].index
    num_questions = len(category_indices)
    plot_data.loc[category_indices, 'y_pos'] = range(current_y, current_y + num_questions)
    current_y += num_questions + gap_size  # Add gap after each category

# Set up the plot
plt.figure(figsize=(14, 10))

# Create a color palette with as many colors as categories
palette = sns.color_palette('pastel', n_colors=len(category_list))

# Create the bar plot using matplotlib directly
ax = plt.gca()

# Plot bars manually to control positions and colors
for idx, row in plot_data.iterrows():
    ax.barh(
        y=row['y_pos'],
        width=row['overall_score_mean'],  # Use 'overall_score_mean' instead of 'mean_of_means'
        xerr=row['overall_score_std'],    # Use 'overall_score_std' instead of 'mean_of_stds'
        height=0.8,
        color=palette[row['category_idx']],
        capsize=3,
        error_kw={'ecolor': 'black'},
        label=row['category_id'] if idx == plot_data[plot_data['category_id'] == row['category_id']].index[0] else ""
    )

# Remove duplicate labels for legend
handles, labels = ax.get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys(), title='Categories', bbox_to_anchor=(1.05, 1), loc='upper left')

# Customize the plot
plt.title('Overall Score Mean with Standard Deviations for Each Question Index Grouped by Category')
plt.xlabel('Overall Score Mean')
# Remove y-axis label as labels are self-explanatory
plt.ylabel('')

# Set y-ticks and labels
plt.yticks(plot_data['y_pos'], plot_data['Category_Question'])

# Invert y-axis so that the first question appears on top
plt.gca().invert_yaxis()

# Show grid lines for clarity
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Adjust layout to avoid overlap
plt.tight_layout()

# Display the plot
plt.show()
```

```python
### Plot 3: Scores per Question for Each Category with Standard Deviation

# Convert 'question_index' to integer for sorting
question_df['question_index'] = question_df['question_index'].astype(int)

# Prepare data for plotting
score_columns = ['Q1_mean', 'Q2_mean', 'Q3_mean', 'Q4_mean', 'Q5_mean']
std_columns = ['Q1_std', 'Q2_std', 'Q3_std', 'Q4_std', 'Q5_std']

# Loop over each category
for category in question_df['category_id'].unique():
    # Filter data for the category
    cat_df = question_df[question_df['category_id'] == category]
    
    # Melt the DataFrame to long format for means
    melted_mean_df = pd.melt(
        cat_df,
        id_vars=['question_index'],
        value_vars=score_columns,
        var_name='Question',
        value_name='Score Mean'
    )
    
    # Melt the DataFrame to long format for standard deviations
    melted_std_df = pd.melt(
        cat_df,
        id_vars=['question_index'],
        value_vars=std_columns,
        var_name='Question',
        value_name='Score Std'
    )
    
    # Clean up the 'Question' column to match
    melted_mean_df['Question'] = melted_mean_df['Question'].str.replace('_mean', '')
    melted_std_df['Question'] = melted_std_df['Question'].str.replace('_std', '')
    
    # Merge the mean and std dataframes
    melted_df = pd.merge(
        melted_mean_df,
        melted_std_df,
        on=['question_index', 'Question']
    )
    
    # Sort data
    melted_df = melted_df.sort_values('question_index').reset_index(drop=True)
    
    # Set up the plot
    plt.figure(figsize=(12, 6))
    
    # Create grouped bar plot
    ax = sns.barplot(
        data=melted_df,
        x='question_index',
        y='Score Mean',
        hue='Question',
        palette='pastel',       # Do not compute confidence intervals
        errorbar=None    # For seaborn version 0.12 and above
    )
    
    # Add error bars
    # Get the positions of the bars
    for (index, row), bar in zip(melted_df.iterrows(), ax.patches):
        # Calculate the center of each bar
        x = bar.get_x() + bar.get_width() / 2
        y = bar.get_height()
        yerr = row['Score Std']
        # Draw the error bar
        ax.errorbar(
            x=x,
            y=y,
            yerr=yerr,
            fmt='none',
            ecolor='black',
            capsize=3
        )
    
    # Add labels and title
    plt.xlabel('Question Index')
    plt.ylabel('Score Mean')
    plt.title(f'Scores per Question for Category {category} with Standard Deviation')
    
    # Adjust legend
    plt.legend(title='Questions', bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Show grid lines for clarity
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Display the plot
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import levene, kruskal

# Load the data
df = pd.read_csv('evaluation_results_GPT4-score.csv')

# Exclude 'overall' rows
df_filtered = df[df['question_index'] != 'overall']

# Convert 'category_id' to a categorical variable
df_filtered['category_id'] = df_filtered['category_id'].astype('category')

# List of question mean columns
question_means = ['Q1_mean', 'Q2_mean', 'Q3_mean', 'Q4_mean', 'Q5_mean']

# Loop over each question mean
for question in question_means:
    print(f"\nAnalysis for {question}:")

    # Perform ANOVA
    model = ols(f'{question} ~ C(category_id)', data=df_filtered).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print("ANOVA results:")
    print(anova_table)

    # Levene's Test for Homogeneity of Variances
    groups = [group[question].dropna().values for name, group in df_filtered.groupby('category_id')]
    stat, p_levene = levene(*groups)
    print(f"Levene's test: p-value = {p_levene}")

    # If Levene's test is significant, perform Kruskal-Wallis Test
    if p_levene < 0.05:
        stat, p_kw = kruskal(*groups)
        print(f"Kruskal-Wallis test: p-value = {p_kw}")
    else:
        print("Variances are equal across groups (Levene's test not significant).")

    # Boxplot
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='category_id', y=question, data=df_filtered, palette='pastel')
    plt.title(f'{question} Scores Across Categories')
    plt.xlabel('Category ID')
    plt.ylabel(f'{question}')
    plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import levene, kruskal
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Step 1: Load the data
df = pd.read_csv('evaluation_results_GPT4-score.csv')

# Step 2: Exclude 'overall' rows and convert 'question_index' to integer
df_filtered = df[df['question_index'] != 'overall'].copy()
df_filtered['question_index'] = df_filtered['question_index'].astype(int)

# Step 3: Create new subcategories within Category 4
def map_category(row):
    if row['category_id'] == '4':
        if 0 <= row['question_index'] <= 5:
            return 'category_4_bias'
        elif 6 <= row['question_index'] <= 10:
            return 'category_4_harmfulness'
    return row['category_id']

# Apply the function to create the new 'category_id_updated' column
df_filtered['category_id_updated'] = df_filtered.apply(map_category, axis=1)

# Update 'category_id' with 'category_id_updated'
df_filtered['category_id'] = df_filtered['category_id_updated']
df_filtered = df_filtered.drop(columns=['category_id_updated'])

# Ensure 'category_id' is treated as a categorical variable
df_filtered['category_id'] = df_filtered['category_id'].astype('category')

# Step 4: Perform ANOVA for each question score
question_means = ['Q1_mean', 'Q2_mean', 'Q3_mean', 'Q4_mean', 'Q5_mean']

for question in question_means:
    print(f"\nAnalysis for {question}:")
    
    # Perform ANOVA
    model = ols(f'{question} ~ C(category_id)', data=df_filtered).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print("ANOVA results:")
    print(anova_table)
    
    # Levene's Test for Homogeneity of Variances
    groups = [group[question].dropna().values for name, group in df_filtered.groupby('category_id')]
    stat, p_levene = levene(*groups)
    print(f"Levene's test: p-value = {p_levene}")
    
    # If Levene's test is significant, perform Kruskal-Wallis Test
    if p_levene < 0.05:
        stat, p_kw = kruskal(*groups)
        print(f"Kruskal-Wallis test: p-value = {p_kw}")
    else:
        print("Variances are equal across groups (Levene's test not significant).")
    
    # Generate boxplot
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='category_id', y=question, data=df_filtered, palette='pastel')
    plt.title(f'{question} Scores Across Categories')
    plt.xlabel('Category ID')
    plt.ylabel(f'{question}')
    plt.show()

# Step 5: Compute correlation matrices for the new subcategories
# Filter data for the new subcategories
subcategories_df = df_filtered[df_filtered['category_id'].isin(['category_4_bias', 'category_4_harmfulness'])]

# Select relevant columns
columns_of_interest = ['Q4_mean', 'Q5_mean', 'overall_score_mean']

# Loop through each subcategory
for subcategory in ['category_4_bias', 'category_4_harmfulness']:
    print(f"\nCorrelation Matrix for {subcategory}:")
    sub_df = subcategories_df[subcategories_df['category_id'] == subcategory]
    data = sub_df[columns_of_interest].dropna()
    correlation_matrix = data.corr()
    print(correlation_matrix)
    
    # Visualize the correlation matrix
    plt.figure(figsize=(8, 6))
    mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', mask=mask, square=True, linewidths=.5)
    plt.title(f'Correlation Matrix for {subcategory}')
    plt.tight_layout()
    plt.show()

# Step 6: Perform t-tests to compare mean scores between the two subcategories
columns_to_compare = ['Q4_mean', 'Q5_mean', 'overall_score_mean']

for column in columns_to_compare:
    print(f"\nT-test for {column} between subcategories:")
    group1 = subcategories_df[subcategories_df['category_id'] == 'category_4_bias'][column].dropna()
    group2 = subcategories_df[subcategories_df['category_id'] == 'category_4_harmfulness'][column].dropna()
    stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)
    print(f"t-statistic = {stat}, p-value = {p_value}")

# Optional: Additional visualizations or analyses can be added as needed
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('./evaluation_results/evaluation_results_merged.csv')
df_rephrased = pd.read_csv('./evaluation_results/evaluation_results_rephrased_check_merged.csv')

# Exclude rows where 'category_id' is '3_MCQ'
df = df[df['category_id'] != '3_MCQ']
df_rephrased = df_rephrased[df_rephrased['category_id'] != '3_MCQ']

# Prepare data for the first subplot
grouped_df = df.groupby('category_id')['f1_score_mean'].agg(['mean', 'std']).reset_index()
grouped_df_rephrased = df_rephrased.groupby('category_id')['f1_score_mean'].agg(['mean', 'std']).reset_index()

# Merge grouped dataframes on 'category_id'
grouped_merged = pd.merge(
    grouped_df,
    grouped_df_rephrased,
    on='category_id',
    suffixes=('_original', '_rephrased')
)

# Prepare data for the second subplot
df_sorted = df.sort_values(by=['category_id', 'question_index']).reset_index(drop=True)
df_rephrased_sorted = df_rephrased.sort_values(by=['category_id', 'question_index']).reset_index(drop=True)

# Merge df_sorted and df_rephrased_sorted on 'category_id' and 'question_index'
merged_df = pd.merge(
    df_sorted[['category_id', 'question_index', 'f1_score_mean']],
    df_rephrased_sorted[['category_id', 'question_index', 'f1_score_mean']],
    on=['category_id', 'question_index'],
    suffixes=('_original', '_rephrased'),
    how='inner'
)

# Initialize variables for the second subplot
y_positions = []
y_labels = []
f1_scores_original = []
f1_scores_rephrased = []
colors_list = []
category_separators = []
category_labels_positions = []
category_labels = []

y_pos = 0  # Starting y position
category_gap = 1  # Gap between categories
colors = plt.cm.tab20.colors  # Colormap for different categories

categories = merged_df['category_id'].unique()

for i, category in enumerate(categories):
    category_data = merged_df[merged_df['category_id'] == category].reset_index(drop=True)
    num_questions = len(category_data)
    
    # Store position for category label (middle of the category block)
    category_labels_positions.append(y_pos + (num_questions - 1) / 2)
    category_labels.append(f"Category {category}")
    
    # Assign y positions to questions within the category
    for idx in range(num_questions):
        y_positions.append(y_pos)
        y_labels.append(f"Q{int(category_data.iloc[idx]['question_index'])}")
        f1_scores_original.append(category_data.iloc[idx]['f1_score_mean_original'])
        f1_scores_rephrased.append(category_data.iloc[idx]['f1_score_mean_rephrased'])
        colors_list.append(colors[i % len(colors)])
        y_pos += 1  # Move to next position
    
    # Add a separator after each category except the last one
    if i < len(categories) - 1:
        category_separators.append(y_pos - 0.5)
        y_pos += category_gap  # Add gap before the next category

# Set up the figure and axes
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 18))
plt.subplots_adjust(hspace=0.5)

# First subplot: Average F1 Score by Category ID
# Plot rephrased data
axes[0].barh(
    y=grouped_merged['category_id'],
    width=grouped_merged['mean_rephrased'],
    color='white',
    edgecolor='black',
    label='Rephrased'
)

# Plot original data
axes[0].barh(
    y=grouped_merged['category_id'],
    width=grouped_merged['mean_original'],
    xerr=grouped_merged['std_original'],
    color='skyblue',
    ecolor='gray',
    capsize=5,
    label='Original'
)

axes[0].set_xlabel('Average F1 Score')
axes[0].set_ylabel('Category ID')
axes[0].set_title('Average F1 Score by Category ID with Standard Deviation')
axes[0].invert_yaxis()  # Highest category at the top
axes[0].legend()

# Second subplot: F1 Score by Question Index and Category ID
# Plot rephrased data bars (background bars)
axes[1].barh(
    y=y_positions,
    width=f1_scores_rephrased,
    color='white',
    edgecolor='black',
    label='Rephrased'
)

# Plot original data bars (foreground bars)
axes[1].barh(
    y=y_positions,
    width=f1_scores_original,
    color=colors_list,
    edgecolor='black',
    label='Original'
)

# Set y-axis ticks and labels
axes[1].set_yticks(y_positions)
axes[1].set_yticklabels(y_labels)

# Add category separators
for sep in category_separators:
    axes[1].axhline(y=sep, color='black', linewidth=1)

# Add category labels on the left side
for pos, label in zip(category_labels_positions, category_labels):
    axes[1].text(
        x=-0.08,  # Slightly left of the y-axis (adjust as needed)
        y=pos,
        s=label,
        va='center',
        ha='right',
        fontsize=12,
        transform=axes[1].get_yaxis_transform()
    )

axes[1].set_xlabel('F1 Score')
axes[1].set_ylabel('Questions')
axes[1].set_title('F1 Score by Question Index and Category ID')
axes[1].invert_yaxis()
axes[1].legend()

# Adjust layout and show plot
plt.tight_layout()
plt.savefig('f1_scores.jpg')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
df = pd.read_csv('./evaluation_results/evaluation_results_merged.csv')
df_rephrased = pd.read_csv('./evaluation_results/evaluation_results_rephrased_check_merged.csv')

# Exclude rows where 'category_id' is '3_MCQ'
df = df[df['category_id'] != '3_MCQ']
df_rephrased = df_rephrased[df_rephrased['category_id'] != '3_MCQ']

# Prepare data for the first subplot
grouped_df = df.groupby('category_id')['f1_score_mean'].agg(['mean', 'std']).reset_index()
grouped_df_rephrased = df_rephrased.groupby('category_id')['f1_score_mean'].agg(['mean', 'std']).reset_index()

# Merge grouped dataframes on 'category_id'
grouped_merged = pd.merge(
    grouped_df,
    grouped_df_rephrased,
    on='category_id',
    suffixes=('_original', '_rephrased')
)

# Prepare data for the second subplot
df_sorted = df.sort_values(by=['category_id', 'question_index']).reset_index(drop=True)
df_rephrased_sorted = df_rephrased.sort_values(by=['category_id', 'question_index']).reset_index(drop=True)

# Merge df_sorted and df_rephrased_sorted on 'category_id' and 'question_index'
merged_df = pd.merge(
    df_sorted[['category_id', 'question_index', 'f1_score_mean']],
    df_rephrased_sorted[['category_id', 'question_index', 'f1_score_mean']],
    on=['category_id', 'question_index'],
    suffixes=('_original', '_rephrased'),
    how='inner'
)

# Initialize variables for the second subplot
y_positions_original = []
y_positions_rephrased = []
y_labels = []
f1_scores_original = []
f1_scores_rephrased = []
colors_list = []
category_separators = []
category_labels_positions = []
category_labels = []

y_pos = 0  # Starting y position
question_gap = 1.0  # Space between questions
category_gap = 2.0  # Gap between categories
bar_height = 0.4  # Height of each bar
colors = plt.cm.tab20.colors  # Colormap for different categories

categories = merged_df['category_id'].unique()

for i, category in enumerate(categories):
    category_data = merged_df[merged_df['category_id'] == category].reset_index(drop=True)
    num_questions = len(category_data)
    
    # Calculate position for category label
    start_pos = y_pos
    end_pos = y_pos + (num_questions * question_gap)
    category_labels_positions.append((start_pos + end_pos - question_gap) / 2)
    category_labels.append(f"Category {category}")
    
    # Assign y positions to questions within the category
    for idx in range(num_questions):
        # Positions for original and rephrased bars
        y_positions_original.append(y_pos)
        y_positions_rephrased.append(y_pos + bar_height )  # Small offset to separate bars
        y_labels.append(f"Q{int(category_data.iloc[idx]['question_index'])}")
        f1_scores_original.append(category_data.iloc[idx]['f1_score_mean_original'])
        f1_scores_rephrased.append(category_data.iloc[idx]['f1_score_mean_rephrased'])
        colors_list.append(colors[i % len(colors)])
        y_pos += question_gap  # Move to next position
    
    # Add a separator after each category except the last one
    if i < len(categories) - 1:
        category_separators.append(y_pos - question_gap / 2)
        y_pos += category_gap - question_gap  # Add gap before the next category

# Set up the figure and axes
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 18))
plt.subplots_adjust(hspace=0.5)

# First subplot: Average F1 Score by Category ID
# Plot rephrased data
axes[0].barh(
    y=grouped_merged['category_id'],
    width=grouped_merged['mean_rephrased'],
    color='white',
    edgecolor='black',
    label='Rephrased'
)

# Plot original data
axes[0].barh(
    y=grouped_merged['category_id'],
    width=grouped_merged['mean_original'],
    xerr=grouped_merged['std_original'],
    color='skyblue',
    ecolor='gray',
    capsize=5,
    label='Original'
)

axes[0].set_xlabel('Average F1 Score')
axes[0].set_ylabel('Category ID')
axes[0].set_title('Average F1 Score by Category ID with Standard Deviation')
axes[0].invert_yaxis()  # Highest category at the top
axes[0].legend()

# Second subplot: F1 Score by Question Index and Category ID
# Plot rephrased data bars
axes[1].barh(
    y=y_positions_rephrased,
    width=f1_scores_rephrased,
    height=bar_height,
    color='white',
    edgecolor='black',
    label='Rephrased'
)

# Plot original data bars
axes[1].barh(
    y=y_positions_original,
    width=f1_scores_original,
    height=bar_height,
    color=colors_list,
    edgecolor='black',
    label='Original'
)

# Set y-axis ticks and labels (only for original data positions)
axes[1].set_yticks(y_positions_original)
axes[1].set_yticklabels(y_labels)

# Add category separators
for sep in category_separators:
    axes[1].axhline(y=sep, color='black', linewidth=1)

# Add category labels on the left side
for pos, label in zip(category_labels_positions, category_labels):
    axes[1].text(
        x=-0.02,  # Slightly left of the y-axis (adjust as needed)
        y=pos,
        s=label,
        va='center',
        ha='right',
        fontsize=12,
        transform=axes[1].get_yaxis_transform()
    )

axes[1].set_xlabel('F1 Score')
axes[1].set_ylabel('Questions')
axes[1].set_title('F1 Score by Question Index and Category ID')
axes[1].invert_yaxis()
axes[1].legend()

# Adjust layout and show plot
plt.tight_layout()
plt.savefig('f1_scores.jpg')
plt.show()
```
